```python
import sys
sys.path.append("../")
from bandipy import simulation
import numpy as np
## For synthetic data generation
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from keras import backend as K
```

### Note:
`ads_f_name` is the address (directory+filename) of the created dataset that you chose in`create_datasets.ipynb`. If you do not use our default `train_1000000000_rows`, then you also need to set `ads_build` to `True` for the first time.

```python
context_size = 10
n_actions = 40
early_frac = .70
time_rng = np.arange(20,301,20)
n_users = 3000
reports = list()
with_data = False
real_contexts = None
real_responses = None
for n_samples in time_rng:
    
    print("___________________"+str(n_samples)+"_______________________________")
 
    sim = simulation.Simulation(data_type='ads', 
                            bandit_algorithm='contextual_linear_ucb',
                            privacy_model = 'crowd_blending_with_sampling', 
                            sim_sig = 'tmp')

    report, real_contexts, real_responses = sim.run_simulation(
                   n_users=n_users, early_frac=early_frac, 
                   n_samples=n_samples,
                   n_actions=n_actions, context_size=context_size,
                   ads_dataset = 'criteo_kaggle',
                   ads_f_name = "train_1000000000_rows",
                   ads_build = False,
                   alpha =1.,
                   cb_sampling_rate = .5,
                   neg_rew_sam_rate = .05,
                   cb_context_threshold = 10,
                   dec_digits = 1,
                   bin_size = 5,
                   with_data = with_data,
                   contexts = real_contexts,
                   responses = real_responses
                   )
    with_data = True

    reports.append(report)
    np.save("_SySML_Rep_CKA_"+str(n_users)+"_"+str(n_actions)+"_"+str(context_size)+".npy", np.array(reports))
    print(np.array(reports)) 
reports = np.array(reports)
```

```python
import matplotlib.pyplot as plt
rng = np.arange(20,301,20)
def paint_reports(reports, rng):
    x_ax = rng
    fmts = ['bs-', 'ko-', 'r*-']
    plt.errorbar(x_ax, reports[:,0,0], reports[:,0,1]//1, fmt=fmts[0], ms=15, ecolor="blue", label="Cold")
    plt.errorbar(x_ax, reports[:,1,0], reports[:,1,1]//1, fmt=fmts[1], ms=15, ecolor="black", label="Warm_Non_Private")
    plt.errorbar(x_ax, reports[:,2,0], reports[:,2,1]//1, fmt=fmts[2], ms=20, ecolor="red", label="Warm_Private")

# users_rng = np.array([1000., 5000., 10000., 50000., 100000.])
plt.figure(figsize=(14,6))
paint_reports(reports, rng)
# plt.xscale('log')
plt.xlabel('Number of Observed Samples by Local Agents', size=20)
plt.ylabel('Click Through Rate', size=20)
plt.xticks(rng,size=20)
plt.yticks(size=20)
plt.legend( ncol=1, prop={'size': 18})
plt.show()
```
