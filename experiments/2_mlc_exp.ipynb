```python
import sys
sys.path.append("../")
from bandipy import simulation
import numpy as np
## For synthetic data generation
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from keras import backend as K
```

```python
context_size = 20
#n_actions = 40 #mm
n_actions = 20 #tmc
early_frac = .70
time_rng = np.append([2],np.arange(5,101,5))
# n_users = 425 #mm
n_users = 275 #tmc
reports = list()

with_data = False
real_contexts = None
real_responses = None
mlc_dataset = "tmc"
for n_samples in time_rng:
    
    print("___________________"+str(n_samples)+"_______________________________")
 
    sim = simulation.Simulation(data_type='mlc', 
                            bandit_algorithm='contextual_linear_ucb',
                            privacy_model = 'crowd_blending_with_sampling', 
                            sim_sig = 'tmp')

    report, real_contexts, real_responses = sim.run_simulation(n_users=n_users, early_frac=early_frac,
                                                               n_samples=n_samples,
                                                               n_actions=n_actions, context_size=context_size,
                                                               with_data = with_data,
                                                               mlc_dataset = mlc_dataset,
                                                               alpha =1.,
                                                               cb_sampling_rate = .5,
                                                               neg_rew_sam_rate = .05,
                                                               cb_context_threshold = 10,
                                                               dec_digits = 1,
                                                               bin_size = 5,
                                                               contexts = real_contexts,
                                                               responses = real_responses
                                                               )
    with_data = True

    reports.append(report)
    np.save("SySML_"+str(mlc_dataset)+"_"+str(n_users)+"_"+str(n_actions)+"_"+str(context_size)+".npy", np.array(reports))
    print(np.array(reports)) 
reports = np.array(reports)
```

```python
import matplotlib.pyplot as plt
def paint_reports(reports, time_rng):
    x_ax = time_rng
    fmts = ['bs-', 'ko-', 'r*-']
    plt.errorbar(x_ax, reports[:,0,0]*100, reports[:,0,1]//10, fmt=fmts[0], ms=15, ecolor="blue", label="Cold")
    plt.errorbar(x_ax, reports[:,1,0]*100, reports[:,1,1]//10, fmt=fmts[1], ms=15, ecolor="black", label="Warm_Non_Private")
    plt.errorbar(x_ax, reports[:,2,0]*100, reports[:,2,1]//10, fmt=fmts[2], ms=20, ecolor="red", label="Warm_Private")

plt.figure(figsize=(14,8))
paint_reports(reports, time_rng)
plt.xlabel('Number of Observed Samples by Local Agents', size=20)
plt.ylabel('Accuracy (%)', size=20)
plt.xticks(np.arange(10,101,10),size=20)
plt.yticks(size=20)
plt.legend( ncol=1, prop={'size': 18})
plt.show()
```
